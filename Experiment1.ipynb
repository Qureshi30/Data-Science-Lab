```python
import pandas as pd

import matplotlib.pyplot as plt
df = pd.read_csv('titanic.csv')
```

```python
print(df.head())
print(df.isnull().sum())
print(df['age'].mean())
print(df['age'].mean(skipna=True))

counts = df["survived"].value_counts()
print("Not survived:", counts[0])
print("Survived:", counts[1])
```

```python
fare_sorted = df["fare"].sort_values().reset_index(drop=True)

plt.step(range(len(fare_sorted)), fare_sorted)
plt.title("Step Plot: Sorted Fare")
plt.xlabel("Passenger Index (sorted)")
plt.ylabel("Fare")
plt.show()
```

```python
plt.scatter(df["age"], df["fare"])
plt.title("Scatter Plot: Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python

table = pd.crosstab(df["who"], df["survived"])

ax = table.plot(kind="bar", figsize=(8,5))

# Add labels with padding
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=10)

# Increase y-limit so small bars labels are visible
ax.set_ylim(0, table.values.max() + 50)

plt.title("Survival by Who (Counts Visible)")
plt.xlabel("Who")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(["Dead (0)", "Survived (1)"])
plt.show()
```

```python
age = df["age"].dropna().sort_values().reset_index(drop=True)

plt.plot(age, label="Age")
plt.fill_between(range(len(age)), age, alpha=0.5)
plt.title("Fill Between: Age Values")
plt.xlabel("Index (sorted)")
plt.ylabel("Age")
plt.show()
```

```python
plt.boxplot(df["fare"].dropna())
plt.title("Box Plot: Fare")
plt.ylabel("Fare")
plt.show()
```

```python
plt.hist(df["age"].dropna(), bins=20)
plt.title("Histogram: Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
